Below, a dataset with measurements of cancerous and non-cancerous breast tumors is loaded and displayed.  Use `LogisticRegression` and `KNeighborsClassifier` to build predictive models on train/test splits.  Generate a confusion matrix and explore the classifiers mistakes.  

- Which model do you prefer and why?

```python
cancer = load_breast_cancer(as_frame=True).frame
```

```python
cancer.head()
```

```python
cancer['target'] = 1 - cancer['target']
```

```python
cancer.head()
```

```python
from sklearn.model_selection import train_test_split
from re import X
```

```python
#splitting the data
X = cancer.iloc[:, :-1].values # Select all rows and all columns except the last one (target)
y = cancer['target'].values # Extract the target column as values

X_train, X_test, y_train, y_test = train_test_split(X, y) # Now this should
```

```python
lgr = LogisticRegression()
knn = KNeighborsClassifier()
#instantiating the matrix
```

```python
scaler = StandardScaler()
#convert data to z scores
```

```python
from sklearn.pipeline import Pipeline
```

```python
# Corrected Pipeline definitions:
lgr_pipe = Pipeline([('scale', scaler), ('model', lgr)])
knn_pipe = Pipeline([('scale', scaler), ('model', knn)])

# Now, you can use the fit method:
lgr_pipe.fit(X_train, y_train)
knn_pipe.fit(X_train, y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

# Logistic Regression Confusion Matrix
y_pred_lgr = lgr_pipe.predict(X_test)  # Predict using the logistic regression model
cm_lgr = confusion_matrix(y_test, y_pred_lgr)
disp_lgr = ConfusionMatrixDisplay(confusion_matrix=cm_lgr, display_labels=lgr.classes_)
disp_lgr.plot()
plt.title("Logistic Regression Confusion Matrix")
plt.show()


# KNN Confusion Matrix
y_pred_knn = knn_pipe.predict(X_test)  # Predict using the knn model
cm_knn = confusion_matrix(y_test, y_pred_knn)
disp_knn = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=knn.classes_)
disp_knn.plot()
plt.title("KNN Confusion Matrix")
plt.show()
```

TP/FN+TP = Recall

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lgr_pipe, X_train, y_train)
```

```python
cross_val_score(lgr_pipe, X_train, y_train, scoring='recall')
```

### Problem

Below, a dataset around customer churn is loaded and displayed.  Build classification models on the data and visualize the confusion matrix.  

- Suppose you want to offer an incentive to customers you think are likely to churn, what is an appropriate evaluation metric?
- Suppose you only have a budget to target 100 individuals you expect to churn.  By targeting the most likely predictions to churn, what percent of churned customers did you capture?

```python
churn = fetch_openml(data_id = 43390).frame
```

```python
churn.head()
```

```python
X = churn.iloc[:, :-1].values
y = churn['Exited'].values
```

```python
lgrs = LogisticRegression()
knns = KNeighborsClassifier()
```

```python
scaler = StandardScaler()
```

```python
# Identify categorical columns - these columns likely contain strings
categorical_cols = churn.select_dtypes(include=['object']).columns
```

```python
# Create dummy variables for categorical features
churn_encoded = pd.get_dummies(churn, columns=categorical_cols, drop_first=True)
```

```python
X = churn_encoded.drop('Exited', axis=1).values  # Assuming 'Exited' is your target column
y = churn_encoded['Exited'].values

# Proceed with the rest of your code
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=11)
```

```python
# Corrected Pipeline definitions:
lgrs_pipe = Pipeline([('scale', scaler), ('model', lgrs)])
knns_pipe = Pipeline([('scale', scaler), ('model', knns)])


# Now, you can use the fit method:
lgrs_pipe.fit(X_train, y_train)
knns_pipe.fit(X_train, y_train)
```

```python
y_pred_lgr = lgrs_pipe.predict(X_test)  # Predict using the logistic regression model
cm_lgr = confusion_matrix(y_test, y_pred_lgr)


disp_lgr = ConfusionMatrixDisplay(confusion_matrix=cm_lgr, display_labels=lgrs.classes_)
disp_lgr.plot()
plt.title("Logistic Regression Confusion Matrix")
plt.show()


# KNN Confusion Matrix
y_pred_knn = knns_pipe.predict(X_test)  # Predict using the knn model
cm_knn = confusion_matrix(y_test, y_pred_knn)
disp_knn = ConfusionMatrixDisplay(confusion_matrix=cm_knn, display_labels=knns.classes_)
disp_knn.plot()
plt.title("KNN Confusion Matrix")
plt.show()
```

```python
lgrs_pipe.predict_proba(X_test)
#seeing the probability of what classified as 0'S AND 1'S
```

```python
np.sort(lgrs_pipe.predict_proba(X_test)[:, 1])
#SORT THE ONE'S PROBABILITY
```
