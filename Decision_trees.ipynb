### Entropy

One way to quantify the quality of the split is to use a quantity called **entropy**.  This is determined by:

$$H = - \sum p_i \log p_i $$

With a decision tree the idea is to select a feature that produces less entropy.  

Information Gain: Entropy is often used in decision trees to calculate information gain, which helps in selecting the best attribute for splitting the data. Lower entropy means higher information gain.

Uncertainty in Predictions: High entropy indicates that the data or predictions are highly uncertain (e.g., a 50-50 distribution of classes in a binary classification). Low entropy suggests more certainty.

Data Distribution: It helps in understanding how well-distributed or imbalanced the data is across different classes or categories.

p(x) is the probability of class

Example:
In a binary classification:

Low Entropy: Class distribution of [90%, 10%] (mostly one class).
High Entropy: Class distribution of [50%, 50%] (more uncertainty).
Understanding entropy is crucial for optimizing machine learning models, particularly in feature selection and model performance analysis.

```python
#all the same -- probability = 1
1*np.log2(1)
```

```python
#half and half -- probability = .5
-(1/2*np.log2(1/2) + 1/2*np.log2(1/2))
```

```python
#subset the data to age, pclass, and survived five rows
data = titanic.loc[:4, ['age', 'pclass', 'survived']]
data
```

```python
#compute entropy for pclass
#first class entropy
first_class_entropy = -np.sum(2/5 *(1 * np.log2(1)) + 3/5 * (1/3* np.log2(1/3)))
first_class_entropy
```

```python
#pclass entropy
third_class_entropy = ''
```

```python
#weighted sum of these
pclass_entropy = ''
```

```python
#splitting on age < 30
entropy_left = -(1/2*np.log2(1/2) + 1/2*np.log2(1/2))
entropy_right = -(1/3*np.log2(1/3) + 2/3*np.log2(2/3))
entropy_age = 2/5*entropy_left + 3/5*entropy_right
entropy_age
```

```python
#original entropy
original_entropy = -((3/5)*np.log2(3/5) + (2/5)*np.log2(2/5))
original_entropy
```

```python
# improvement based on pclass
original_entropy - pclass_entropy
```

```python
#improvement based on age < 30
original_entropy - entropy_age
```
